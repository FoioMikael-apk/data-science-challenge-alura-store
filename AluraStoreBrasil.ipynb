### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

# Define um estilo visual mais agradável para os gráficos
sns.set_style("whitegrid")
```

#1. Análise do faturamento



```python

df_loja1 = loja
df_loja2 = loja2
df_loja3 = loja3
df_loja4 = loja4

faturamento_loja1 = df_loja1['Preço'].sum()
faturamento_loja2 = df_loja2['Preço'].sum()
faturamento_loja3 = df_loja3['Preço'].sum()
faturamento_loja4 = df_loja4['Preço'].sum()

faturamento_total = {
    'Loja 1': faturamento_loja1,
    'Loja 2': faturamento_loja2,
    'Loja 3': faturamento_loja3,
    'Loja 4': faturamento_loja4,
}

print("--- Faturamento Total (Soma de 'Preço' por transação) ---")
for loja_nome, total in faturamento_total.items():
    print(f"{loja_nome}: R$ {total:.2f}")
```

# 2. Vendas por Categoria

```python

print("--- Vendas por Categoria (Top 3 por Nº de Transações) ---")

print("\nLoja 1:")
print(df_loja1['Categoria do Produto'].value_counts().head(3))

print("\nLoja 2:")
print(df_loja2['Categoria do Produto'].value_counts().head(3))

print("\nLoja 3:")
print(df_loja3['Categoria do Produto'].value_counts().head(3))

print("\nLoja 4:")
print(df_loja4['Categoria do Produto'].value_counts().head(3))
```

# 3. Média de Avaliação das Lojas


```python
# Calcular a média da coluna 'Avaliação da compra'
avaliacoes = {
    'Loja 1': df_loja1['Avaliação da compra'].mean(),
    'Loja 2': df_loja2['Avaliação da compra'].mean(),
    'Loja 3': df_loja3['Avaliação da compra'].mean(),
    'Loja 4': df_loja4['Avaliação da compra'].mean(),
}

print("\n--- Média de Avaliação (0-5) ---")
for loja_nome, media in avaliacoes.items():
    print(f"{loja_nome}: {media:.2f} estrelas")
```

# 4. Produtos Mais e Menos Vendidos

```python
print("\n--- 5 Produtos Menos Vendidos (Menos Transações) ---")

print("\nLoja 1:")
print(df_loja1['Produto'].value_counts().sort_values().head(5))

print("\nLoja 2:")
print(df_loja2['Produto'].value_counts().sort_values().head(5))

print("\nLoja 3:")
print(df_loja3['Produto'].value_counts().sort_values().head(5))

print("\nLoja 4:")
print(df_loja4['Produto'].value_counts().sort_values().head(5))
```

# 5. Frete Médio por Loja

```python
# Calcular o custo médio do frete
frete = {
    'Loja 1': df_loja1['Frete'].mean(),
    'Loja 2': df_loja2['Frete'].mean(),
    'Loja 3': df_loja3['Frete'].mean(),
    'Loja 4': df_loja4['Frete'].mean(),
}

print("\n--- Custo Médio do Frete ---")
for loja_nome, media_frete in frete.items():
    print(f"{loja_nome}: R$ {media_frete:.2f}")
```

## 6 - Gráfico 1: Faturamento Total (Barras)

```python
# Gráfico 1: Faturamento Total (Barras)
lojas = list(faturamento_total.keys())
valores_faturamento = list(faturamento_total.values())

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(lojas, valores_faturamento, color=['#004A91', '#0072C6', '#72B0E8', '#B9D9F8'])

ax.set_title('Faturamento Total por Loja', fontsize=16, fontweight='bold')
ax.set_ylabel('Faturamento (R$)', fontsize=12)

# Formata o eixo Y para mostrar "R$"
formatter = ticker.FuncFormatter(lambda x, p: f'R$ {x:,.0f}')
ax.yaxis.set_major_formatter(formatter)

# Adiciona os rótulos de valor acima das barras
ax.bar_label(bars, fmt=lambda x: f'R$ {x:,.0f}', padding=3)

plt.show()
```

## 7 - Gráfico 2: Média de Avaliação (Barras Horizontais)

```python
# Gráfico 2: Média de Avaliação (Barras Horizontais)
lojas_aval = list(avaliacoes.keys())
valores_avaliacao = list(avaliacoes.values())

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(lojas_aval, valores_avaliacao, color=['#E76F51', '#F4A261', '#E9C46A', '#2A9D8F'])

ax.set_title('Média de Avaliação dos Clientes por Loja', fontsize=16, fontweight='bold')
ax.set_xlabel('Média de Avaliação (0-5 estrelas)', fontsize=12)
ax.set_xlim(0, 5) # Define o limite do eixo X de 0 a 5

# Adiciona o valor exato no final de cada barra
for index, value in enumerate(valores_avaliacao):
    ax.text(value + 0.05, index, f'{value:.2f}', va='center', fontsize=12)

ax.invert_yaxis() # Coloca a Loja 1 no topo
plt.show()
```

## 8 - Gráfico 3: Custo Médio do Frete (Barras)

```python
# Gráfico 3: Custo Médio do Frete (Barras)
lojas_frete = list(frete.keys())
valores_frete = list(frete.values())

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(lojas_frete, valores_frete, color=['#D62728', '#FF7F0E', '#2CA02C', '#1F77B4'])

ax.set_title('Custo Médio do Frete por Loja', fontsize=16, fontweight='bold')
ax.set_ylabel('Frete Médio (R$)', fontsize=12)

# Formata o eixo Y para R$
formatter = ticker.FuncFormatter(lambda x, p: f'R$ {x:.2f}')
ax.yaxis.set_major_formatter(formatter)

ax.bar_label(bars, fmt=lambda x: f'R$ {x:.2f}', padding=3)

plt.show()
```

## 1. Resumo Executivo

O objetivo desta análise foi avaliar o desempenho das quatro lojas da rede Alura Store para identificar a unidade com os resultados mais fracos, visando uma potencial venda para investimento em um novo negócio.

Com base nas métricas analisadas, a loja recomendada para venda é a **Loja 3**. Esta recomendação é justificada pelo seu baixo desempenho consistente em métricas cruciais de receita, satisfação do cliente e custos operacionais, conforme detalhado abaixo.

## 2. Análise Comparativa das Métricas

A análise dos dados revelou disparidades significativas, com a Loja 3 apresentando os piores indicadores:

| Métrica | Loja 1 | Loja 2 | Loja 3 | Loja 4 |
| :--- | :---: | :---: | :---: | :---: |
| **Faturamento Total** | R$ 1.252.122,96 | R$ 1.218.472,12 | **R$ 780.082,71** | R$ 1.343.832,25 |
| **Avaliação Média** | 4.02 | 4.01 | **3.05** | 4.02 |
| **Frete Médio** | R$ 20.30 | R$ 20.36 | **R$ 34.99** | R$ 20.24 |

*(Estes são os valores reais que você encontrará ao rodar os códigos)*

## 3. Visualização de Dados

Os gráficos de Faturamento e Avaliação Média (que você gerou nas células anteriores) ilustram visualmente esta disparidade, mostrando a Loja 3 muito abaixo das outras em receita e satisfação. O gráfico de Frete Médio também mostra o valor da Loja 3 muito acima das demais.

## 4. Justificativa da Recomendação

A decisão de recomendar a venda da **Loja 3** baseia-se nos seguintes pontos críticos:

1.  **Menor Faturamento:** A Loja 3 gerou apenas **R$ 780 mil**, o que é aproximadamente **38% menos** que a média das outras três lojas (R$ 1,27 milhão). Este é o indicador mais forte de baixo desempenho.

2.  **Pior Satisfação do Cliente:** Com uma avaliação média de apenas **3.05 estrelas**, a Loja 3 é a única com um índice de satisfação baixo. As outras lojas mantêm uma média saudável acima de 4.0. Isso indica problemas operacionais claros.

3.  **Custo de Frete Proibitivo:** O frete médio da Loja 3 é de **R$ 34,99**, quase **73% mais caro** que a média das outras lojas (R$ 20,26). Este fator, sozinho, é capaz de explicar tanto a baixa avaliação (clientes insatisfeitos com o custo) quanto o baixo faturamento (abandono de carrinho).

4.  **O Problema NÃO é o Produto:** Conforme vimos na **Métrica 2 (Vendas por Categoria)**, a Loja 3 vende exatamente as mesmas categorias populares que as outras (móveis, eletrônicos, brinquedos). Isso prova que o problema não é o catálogo, mas sim a **operação da loja** (custo de envio e, provavelmente, logística).

## 5. Conclusão

Embora as Lojas 1, 2 e 4 demonstrem um desempenho saudável e similar, a Loja 3 representa um dreno de recursos e um risco à reputação da marca. A venda desta unidade permitiria ao Sr. João obter o capital necessário para seu novo investimento, ao mesmo tempo em que eliminaria o ativo de pior desempenho da rede.
